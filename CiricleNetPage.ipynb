```python
import csv
import random
import string

FIRST_NAMES = [
    "James","Michael","William","Benjamin","Daniel","Matthew","Alexander","Henry","Joseph","Samuel",
    "Emma","Olivia","Ava","Sophia","Isabella","Mia","Charlotte","Amelia","Evelyn","Abigail",
    "Lucas","Mason","Ethan","Noah","Logan","Elijah","Jacob","Jackson","Aiden","Sebastian",
    "Emily","Harper","Ella","Grace","Chloe","Lily","Hannah","Aria","Scarlett","Victoria",
    "Andrew","Christopher","Joshua","David","Ryan","Nathan","Isaac","Caleb","Owen","Julian",
    "Levi","Aaron","Adrian","Thomas","Charles","Eli","Jonathan","Hunter","Connor","Dominic",
    "Zoe","Nora","Layla","Riley","Luna","Aurora","Penelope","Stella","Violet","Hazel",
    "Ellie","Madison","Natalie","Savannah","Brooklyn","Leah","Audrey","Claire","Skylar","Lucy"
]

LAST_NAMES = [
    "Smith","Johnson","Williams","Brown","Jones","Garcia","Miller","Davis","Rodriguez","Martinez",
    "Hernandez","Lopez","Gonzalez","Wilson","Anderson","Thomas","Taylor","Moore","Jackson","Martin",
    "Lee","Perez","Thompson","White","Harris","Sanchez","Clark","Ramirez","Lewis","Walker"
    "King","Wright","Scott","Green","Baker","Adams","Nelson","Hill","Campbell","Mitchell",
    "Roberts","Carter","Phillips","Evans","Turner","Torres","Parker","Collins","Edwards","Stewart",
    "Flores","Morris","Nguyen","Murphy","Rivera","Cook","Rogers","Morgan","Peterson","Cooper",
    "Reed","Bailey","Bell","Gomez","Kelly","Howard","Ward","Cox","Diaz","Richardson"
]

JOB_TITLES = [
    "Data Scientist","Software Engineer","Data Analyst","ML Researcher","Business Analyst",
    "Cloud Engineer","UX Designer","Product Manager","Marketing Analyst","Financial Analyst",
    "AI Engineer","Cyber Security","Systems Engineer","Full StackDev","Solutions Architect"
]

HOBBIES = [
    "Reading Books","Basket ball","Photography","Music Production","Mountain Hiking",
    "Yoga Fitness","Digital Art","Video Gaming","Cooking Recipes","Travel Exploring",
    "Swimming","Creative Writing","Dancing","Cycling Tours","Meditation","Coding Projects",
    "Fitness Training","Nature Hiking","Film Making","Blog Writing"
]

def fix_length(text, min_len, max_len):
    if len(text) < min_len:
        text += ''.join(random.choices(string.ascii_lowercase, k=min_len - len(text)))
    return text[:max_len]

def generate_csv(file="CircleNetPage.csv", n=200000):
    with open(file, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)

        for i in range(1, n + 1):
            name = random.choice(FIRST_NAMES) + random.choice(LAST_NAMES)
            nickname = fix_length(name.lower(), 10, 20)

            job = fix_length(random.choice(JOB_TITLES), 10, 20)
            region = random.randint(1, 50)
            hobby = fix_length(random.choice(HOBBIES), 5, 30)

            writer.writerow([i, nickname, job, region, hobby])

generate_csv()
```
